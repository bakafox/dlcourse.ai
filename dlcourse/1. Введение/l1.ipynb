## Зачем нужно машинное обучение?
Исследование и решение проблем задач, считающихся "на грани" требования человеческих возможностей и невозможности написания точного алгоритма.

Например, такими проблемами являются:
- Распознавание изображений
- Распрознавание голоса
- Машинный перевод
- И даже игры!


## Что же такое машинное обучение?
— это раздел Computer Science, который даёт компьютерам возможность обучаться [решению проблем тип вышеописанных] без необходимости прямого программирования [решению этой проблемы].

## Deep Learning
Отличие подхода Deep Learning заключается в том, что задача выбора признаков (feature engineering) также возложен на алгоритм обучения — то есть мы хотим, чтобы система сама научилась отбирать качественные признаки за нас:

На вход мы даём просто подготовленное изображение, и система сама через много слоёв трансформирует их в некоторый определённый набор простых векторов (то есть какие признаки стоит извлечь из картинки), и как по ним классифицировать это изображение. Результаты получаются намного лучше, чем если бы человек выбирал признаки самостоятельно (и соответсвенно, человеку понять, как именно система выбрала эти признаки, по итогу практически невозможно)!
