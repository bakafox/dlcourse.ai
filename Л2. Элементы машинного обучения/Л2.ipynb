# Повторение: машинное обучение
Машинное обучение — отрасль компьютерных наук, цель которой — наделить компьютер способностями к обучению чему-то, НЕ программируя его этому напрямую.

## Supervised Learning (обучение с учителем)
— один из видов машинного обучения. 

У нас имеется некоторый набор данных (например, картинок) с метками (например, их классы), из которых мы извлекаем некоторые признаки, по которым мы можем их классифицировать. Затем, используя некоторый алгоритм машинного обучения, мы можем классифицировать эти данные по признакам, чтобы "научить" модель примерно разделять данные на классы по признакам и затем использовать для "пресказания" классов новых данных по этим же признакам.

#### То, что модель хорошо приближает точность на тренировочных данных, не означает хорошее предсказание на тестовых данных! Например, метод ближайших соседей даст около-100% точность на train, но покажет низкую точность на test.

## Переобучение и недообучение


**Underfitting** — модель недостаточно сложная, чтобы достаточно хорошо описать разницу между классами.

**Overfitting** — модель слишком сложная и гибкая и подстраивается под любые данные, по сути, запоминая сами тренировочные данные вместо их некоторой генералиазции по признакам, из-за чего плохо распространяя их на тестовые даныне.

"Золотая середина", очевидно, находится между ними двумя. Но как понять, не недообучили или переобучили ли мы модель?

## Валидация и кросс-валидация

Мы можем разделить данные на ТРИ части — тренировочные, валидационные и транировочные.

Идея в том, что мы использем валидационный датасет для подбора гиперпараметров после тренировки на тренировочных данных — и уже затем, после подбора оптимального гиперпараметра, выполняем проверку на тесотвом датасете для замера точности распознавания модели.

То есть, мы используем тестовые данные, чтобы проверить, действительно ли модель хорошо распознаёт/классифицирует/etc. (и насколько), или мы просто подобрали слишком хорошие гиперпараметры на валидационных данных, на которых, в свою очередь мы тестируем их подбор для тренировочных данных.



Мы можем пойти дальше и использовать данные для валидации для тренировки — мы разбиваем тренировочные данные на 5 *(или другое кол-во, обычно берут 5 или 10)* частей, 4 из которых мы используем для тренировки, а 1 — для валидации. После этого мы усредняем полученные после валидации значения гиперпараметров и уже затем делаем замер точности модели на тестовых данных.

Такой способ подбора и проверки оптимальных гиперпараметров называется **кросс-валидация.**
